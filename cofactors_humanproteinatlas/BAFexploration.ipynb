```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import ast
```

```python
cofactorsmapped = pd.read_csv("data/cofactors_mapped_combined.csv").drop("Unnamed: 0", axis = 1)
cofactorsmapped.head()
```

```python
BAF = cofactorsmapped[cofactorsmapped["Complex"] == "BAF"].set_index("Bgee")
BAF.head()
```

```python
#Gene Names 
mapper = cofactorsmapped[["Gene Names", "Bgee"]]
mapper["Gene Names"] = mapper["Gene Names"].apply(lambda x: x.split(" "))
mapper = mapper.explode("Gene Names").rename({"Bgee": "Ensembl", "Gene Names": "Gene Name"}, axis=1)
mapper.head()
```

```python
def mapper_ensembl_getter(string):
    """
    Gets list fo ensembls for specific genes
    """
    filtered = mapper[[string in x for x in mapper["Gene Name"]]]
    return list(filtered["Ensembl"].unique())

mapper_ensembl_getter("BAF53")
```

```python
cofactorspecificity = pd.read_csv("helperdata/cofactor_specificity.csv").drop("Unnamed: 0", axis=1).set_index("Ensembl")
cofactorspecificity.head()
```

```python
BAFspecificity = cofactorspecificity.loc[BAF.index.to_list()].drop_duplicates()
BAFtissuespecificdf = BAFspecificity[BAFspecificity["RNA tissue specificity"] != "Low cell type specificity"]
BAFcellspecificdf = BAFspecificity[BAFspecificity["RNA single cell type specificity"] != "Low cell type specificity"]
BAFnucleispecificdf = BAFspecificity[BAFspecificity["RNA single nuclei brain specificity"] != "Low cell type specificity"]

BAFcellspecificdf.head()
```

```python
def cellspecificity(df):
    df = df[["Ensembl", "Gene", "RNA single cell type specific nTPM"]]

    cellspecific_cells = pd.DataFrame({"Cell Type": [], "Genes": [], "Ensembl": []}).set_index("Cell Type")

    for _, row in df.iterrows():
        celldict = row["RNA single cell type specific nTPM"]
        if pd.notna(celldict):
            cell_dict = ast.literal_eval(celldict)
            for cell, _ in cell_dict.items():
                if cell not in cellspecific_cells.index.to_list():
                    cellspecific_cells.loc[cell] = [[], []]
                cellspecific_cells.at[cell, "Genes"].append(row["Gene"])
                cellspecific_cells.at[cell, "Ensembl"].append(row["Ensembl"])

    return cellspecific_cells
```

```python
cellspecificity(BAFcellspecificdf.reset_index())
```
